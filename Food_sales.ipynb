```python
from google.colab import files
uploaded=files.upload ()


import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Food_Delivery.csv")


df["Total_Amount"] = df["Quantity"] * df["Price"]

print("Food Delivery Analytics")
print(df.head())


revenue = df["Total_Amount"].sum()
print("\nTotal Revenue =", revenue)


popular_food = df.groupby("Food_Item")["Quantity"].sum().sort_values(ascending=False)
print("\nPopular Food Items")
print(popular_food)


peak_orders = df["Order_Hour"].value_counts().sort_values(ascending=False)
print("\nPeak Order Hours")
print(peak_orders)


city_sales = df.groupby("City")["Total_Amount"].sum()
print("\nCity Wise Revenue")
print(city_sales)


ratings = df.groupby("Food_Item")["Rating"].mean()
print("\nFood Ratings")
print(ratings)


plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
popular_food.plot(kind="bar")
plt.title("Popular Foods")

plt.subplot(1,3,2)
city_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("City Revenue")

plt.subplot(1,3,3)
peak_orders.head(5).plot(kind="bar")
plt.title("Peak Order Hours")

plt.tight_layout()
plt.show()
```
