```python
mark1=int(input("enter mark 1:"))
mark2=int(input("enter mark 2:"))
mark3=int(input("enter mark 3:"))
total=mark1+mark2+mark3
average=total/3
print("total:",total)
print("average:",average)
if average>=90:
  print("Grade A")
elif average>=75:
  print("Grade B")
elif average>=50:
  print("Grade C")
else:
  print("fail")
```
