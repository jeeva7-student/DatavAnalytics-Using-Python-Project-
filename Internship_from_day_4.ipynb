Day 7 Task Clean A Messy Sales dataset in phython

```python
import pandas as pd
data={
    "product":["Laptop","mobile","none","tablet","laptop"],
    "sales":[50000,30000,20000,None,50000],
    "quantity":[5,3,None,2,5]
}
df=pd.DataFrame(data)
print("original dataset:")
print(df)
df=df.drop_duplicates()
df["product"]=df["product"].fillna("unknown")
df["sales"]=df["sales"].fillna(df["sales"].mean())
df["quantity"]=df["quantity"].fillna(0)
print("\nCleaned Dataset")
print(df)
```

GroupBy

```python
import pandas as pd
data={
    "Region":["north","south","east","west","north","south"],
    "Product":["laptop","mobile","tablet","laptop","laptop","mobile"],
    "sales":[50000,30000,None,45000,500000,30000]

}
df=pd.DataFrame(data)

region_Sales= df.groupby("Region")["sales"].sum()

print(region_Sales)
print(df.columns)
```

Day 8:-Task Region-Wise Sales Report

```python
import pandas as pd
data={
    "Region":["North","South","North","South","East","West"],
    "Product":["Laptop","Mouse","Keyboard","Laptop","Moniter","Laptop"],
    "Sales":[10000,20000,30000,40000,50000,60000]

}
df=pd.DataFrame(data)
df_active=df[df["Sales"]>0]
df_sorted=df_active.sort_values(["Region","Sales"],ascending=[True,False])
reported=df_sorted.groupby("Region")["Sales"].agg(["sum","mean","max","count"])
print(reported)
```

Monthly Sales Using bar chart and line chart


```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May","June"]
sales=[10000,12000,15000,13000,18000,20000]
plt.figure(figsize=(18,5))
#line chart
plt.subplot(1,3,1)
plt.plot(months,
         sales,
         marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
#Bar Chart
plt.subplot(1,3,2)
plt.bar(months,
        sales,
        color="orange")
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
#PIE CHART
plt.subplot(1,3,3)
expenses=[5000,3000,2000]
categories=["Food","Travel","Shopping"]
plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%",
        colors=["pink","yellow","green"])
plt.title("Expenses distribution")
plt.tight_layout()
plt.show()
```
